#visualization using shipping ecommerce dataset

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd
import numpy as np
```

```python
ship = pd.read_csv('shipping_ecommerce.csv')
ship.head(5)
```

```python
ship['Gender'].value_counts()[0:2]
```

```python
s1_Gen = ship['Gender'].value_counts()[0:2]
```

# transaction based on Gender

```python
plt.pie(s1_Gen,labels=['Female','Male'], startangle = 90,  autopct = "%2.1f%%")
plt.legend()
plt.title("Transaction based on Gender")
plt.show()
```

# Warehouse storage

```python
warehouse=ship['Warehouse_block'].value_counts()[0:6]
warehouse
```

```python
plt.pie(warehouse, labels=['A','B','C','D','F'], startangle = 90,  autopct = "%2.1f%%")
plt.legend(loc='lower right')
plt.title("Warehouse storage")
plt.show()
```

# product based customer rating

```python
ship.groupby('Product_importance')['Customer_rating'].sum().plot(kind='bar',color='pink')
plt.title("product based customer rating")
```

# warehouse storages

```python
ship['Warehouse_block'].value_counts().plot.bar(color=['red','yellow','blue','green','black'])
color=['red','yellow','blue','green','black']

# Adding labels and titles
plt.xlabel('warehouse block')
plt.ylabel('storage')
plt.title('warehouse storage')

# Creating custom legend patches
import matplotlib.patches as mpatches
legend_patches = [mpatches.Patch(color=color, label=block) for color, block in zip(color, ship['Warehouse_block'].value_counts().index)]

# Displaying the legend
plt.legend(handles=legend_patches)
plt.show()
```
